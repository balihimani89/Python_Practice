```python
import numpy as np
```

```python
def heapify(random_array,length,i):
    largest = i 
    l = 2 * i + 1     
    r = 2 * i + 2     
    if l < length and random_array[i] < random_array[l]: 
        largest = l 
    if r < length and random_array[largest] < random_array[r]: 
        largest = r 
    if largest != i: 
        random_array[i],random_array[largest] = random_array[largest],random_array[i]
    
```

```python
random_array=np.random.randint(100,2000,10)
random_array=list(random_array)
```

```python
print(random_array)
length=len(random_array)
for i in range(length//2 - 1, -1, -1):
    heapify(random_array,length,i)
print(random_array)
        
```
