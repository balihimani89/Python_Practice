```python
import numpy as np
```

```python
random_array=[]
```

```python
random_array=np.random.randint(low=0,high=5000,size=1000)
```

```python
random_array
```

```python
decimal_hex={10:'a',11:'b',12:'c',13:'d',14:'e',15:'f'}
DTH=[]
for i in range (0,1000):
    finaloutput=""    
    while random_array[i]>0:
        rem=random_array[i]%16
        if(rem<10):
            rem=str(rem)
        else:
            rem=decimal_hex[rem]
        finaloutput+=rem
        quot=random_array[i]//16
        random_array[i]=quot    
    finaloutput=finaloutput[::-1]
    DTH.append(finaloutput)
    
    
    
```
